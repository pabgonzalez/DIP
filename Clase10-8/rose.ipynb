```python
from PIL import Image, Resampling
import numpy as np
from collections import namedtuple
```

```python
im1024 = Image.open("Fig0219(rose1024).tif")
```

```python
def subsampling(im: Image, shape: tuple) -> Image:
    arr = np.asarray(im)
    x = arr.shape[0] // shape[0]
    y = arr.shape[1] // shape[1]
    new_arr = arr[::x,::y]
    return Image.fromarray(new_arr)
```

```python
Rose = namedtuple(
    typename="Rose",
    field_names=["image", "shape"]
)

imgs = []
for x in range(6):
    shape = 1024 // 2**x
    imgs.append(
        Rose(
            image=subsampling(im1024, (shape, shape)),
            shape=(shape, shape)
        )
    )
```

```python
for rose in imgs:
    print(rose.shape)
    display(rose.image)
```

```python
for rose in imgs:
    print(rose.shape)
    display(rose.image.resize((1024,1024), Resampling.NEAREST))
```
